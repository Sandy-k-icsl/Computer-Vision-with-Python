```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
from PIL import Image
from pylab import *
```

```python
#Show image in grayscale
image = 'coins4.JPG'
img = cv2.imread(image,1)
img_orig = img.copy()
img= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.rcParams['figure.figsize'] =(9,7)
plt.title("Show image in GRAY")
plt.imshow(img,cmap='gray')
plt.show()
```

```python
#Code to blur image using GausianBlur
img =cv2.GaussianBlur(img, (21,21), 0)
plt.rcParams['figure.figsize'] =(9,7)
plt.imshow(img,cmap='gray')
plt.title("Show blur image after using Gaussian Kernel")
plt.show()
thresh = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
cv2.THRESH_BINARY_INV, 11, 1)
plt.imshow(thresh)
plt.title("After Using Adaptive Thresholding")
plt.show()
```

```python
#Detect circular images
circles = cv2.HoughCircles(img,cv2.HOUGH_GRADIENT,0.9, 120,
                            param1=50, param2=20, minRadius=130, maxRadius=190)
circles_rounded = np.uint16(np.around(circles))
```

```python
#Counting the number of coins
image = 'coins4.JPG'
img = cv2.imread(image,1)
img_orig = img.copy()
img= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.rcParams['figure.figsize'] =(9,7)
plt.title("Show 19 coins in GRAY")
plt.imshow(img,cmap='gray')
plt.show()
print(circles_rounded)
print(circles_rounded.shape)
print("I have found " + str(circles_rounded.shape[1]) + " coins")
```

```python
count = 1
for i in circles_rounded[0,:]:
    cv2.circle(img_orig,(i[0],i[1]),i[2],(250,100,100),5)
    cv2.circle(img_orig,(i[0],i[1]),2,(250,200,200),5)
    cv2.putText(img_orig, "Coin " + str(count), (i[0],i[1]), cv2.FONT_HERSHEY_SIMPLEX, 1.1,(250,0,0),2)
    count+=1       
```
