```python
#importing libraries
from PIL import Image
import numpy as np
import matplotlib.pyplot as pl
#convert image to gray
pil_im = Image.open('empire.jpg').convert('L')
#show image
pl.figure()
pl.imshow(pil_im)
pl.gray()
pl.axis('off')
pl.show()
```

```python
#Create Thumbnail
pil_im.thumbnail((128,128))
```

```python
#copy and paste regions
box = (100,100,400,400)
region = pil_im.crop(box)
pl.figure()
pl.imshow(region)
pl.gray()
pl.axis('off')
pl.show()
```

```python
print(region)
```

```python
def showimage(src):
    pl.figure()
    pl.imshow(src)
    pl.axis('off')
    pl.show()
```

```python
showimage(region)
```

```python
#rotate image
region = region.transpose(Image.ROTATE_180)
showimage(region)
```

```python
pil_im.paste(region,box)
showimage(pil_im)
```

```python
out = pil_im.resize((128,128))
showimage(out)
```

```python
out = out.rotate(45)
showimage(out)
```
