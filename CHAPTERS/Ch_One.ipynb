```python
#importing libraries
from PIL import Image
from pylab import *
from numpy import *
from scipy.ndimage import filters
import matplotlib.pyplot as pl

# Read image
im = np.array(Image.open('empire_state_building.jpg'))
print("Shape is: {0} of type {1}".format(im.shape, im.dtype))

# read grayscale version to float array
im = np.array(Image.open('empire_state_building.jpg').convert('L'),'f')
print("Shape is: {0} of type {1}".format(im.shape, im.dtype))

# visualize the pixel value of a small region
col_1, col_2 = 200, 300
row_1, row_2 = 230, 265

# some points
x = [100,100,400,400]
y = [200,500,200,500]

# Crop Image
crop = im[col_1:col_2,row_1:row_2]
cols, rows = crop.shape

print("Created crop of shape: {0}".format(crop.shape))




```

```python
# Plotting Image with axes
pl.figure()
pl.imshow(im)
pl.gray()
pl.plot([row_1, row_2, row_2, row_1, row_1], [col_1, col_1, col_2, col_2, col_1], linewidth=2)
# plot the points 
plot(x,y)
# line plot connecting the first two points
plot(x[:2],y[:2])
pl.title('Plotting: "empire_state_building.jpg"')
pl.show()
```

```python
#The axes are useful for debugging, but if you want a prettier plot add: axis(’off’)
pl.figure()
pl.imshow(im)
pl.gray()
pl.plot([row_1, row_2, row_2, row_1, row_1], [col_1, col_1, col_2, col_2, col_1], linewidth=2)
pl.axis('off')
pl.title('Plotting: "empire_state_building.jpg"')
pl.show()
```
